```python
covid_lung.head() #sanity check to be sure it looks alright
```

```python
import sweetviz as sv #cool package for quick eda purposes

#analyze the dataset
covid_report = sv.analyze(covid_lung)


#display the "report"
covid_report.show_html('COVID_Lung_Data.html')
```
